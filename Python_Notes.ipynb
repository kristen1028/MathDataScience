```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
y = np.array([1, 4, 3, 6, 8, 9, 7, 5, 2, 10])

# Plot with different marker types
plt.figure(figsize=(10, 5))

plt.subplot(231)  # 2x3 grid, 1st subplot
plt.plot(x, y, '.', label='Dot Marker')

plt.subplot(232)  # 2x3 grid, 2nd subplot
plt.plot(x, y, 'o', label='Circle Marker')

plt.subplot(233)  # 2x3 grid, 3rd subplot
plt.plot(x, y, 's', label='Square Marker')

plt.subplot(234)  # 2x3 grid, 4th subplot
plt.plot(x, y, '*', label='Star Marker')

plt.subplot(235)  # 2x3 grid, 5th subplot
plt.plot(x, y, 'x', label='X Marker')

plt.subplot(236)  # 2x3 grid, 6th subplot
plt.plot(x, y, '+', label='Plus Marker')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
y = np.array([1, 4, 3, 6, 8, 9, 7, 5, 2, 10])

markers = ['.', ',', 'o', 'v', '^', '<', '>', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_']

plt.figure(figsize=(12, 8))
for i, marker in enumerate(markers):
    plt.subplot(5, 5, i + 1)
    plt.plot(x, y, marker, label=f'Marker: {marker}')
    plt.title(f'Marker: {marker}')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
M = np.linspace(-10,10,100)
B = np.linspace(-10,10,100)
```

```python
M
```

```python
for m in M:
  for b in B:
    print(m,b)
```

```python
best_loss = 1000000
for m in M:
  for b in B:
    m_guess = m
    b_guess = b
    y_guess = m_guess*x + b_guess

    loss = np.sum(y-y_guess)**2
    if loss < best_loss:
      best_loss = loss

    print(m,b,loss)
```
